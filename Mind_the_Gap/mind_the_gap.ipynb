```python
import pandas as pd
import requests
```

```python
response2008 = []
for i in range(1, 19):
    link = 'http://ergast.com/api/f1/2008/' + str(i) + '/driverStandings.json'
    response2008.append(requests.get(link).json())
lead_score = [0]
HAM_score = [0]
MAS_score = [0,0,0]

for response in response2008:
    lead_score.append(response['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings'][0]['points'])
    for standing in response['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']:
        if standing['Driver']['code'] == "HAM":
            HAM_score.append(standing['points'])
        if standing['Driver']['code'] == "MAS":
            MAS_score.append(standing['points'])


response2008quali = []
for i in range(1, 19):
    link = 'http://ergast.com/api/f1/2008/' + str(i) + '/qualifying.json'
    response2008quali.append(requests.get(link).json())
lead_quali08 = []
HAM_quali = []
MAS_quali = []
for response in response2008quali:
    lead_quali08.append(response['MRData']['RaceTable']['Races'][0]['QualifyingResults'][0]['Q3'])
    for quali in response['MRData']['RaceTable']['Races'][0]['QualifyingResults']:
        if quali['Driver']['code'] == 'MAS':
            if 'Q3' in quali:
                MAS_quali.append(quali['Q3'])
            else:
                MAS_quali.append(100)
        if quali['Driver']['code'] == 'HAM':
            if 'Q3' in quali:
                HAM_quali.append(quali['Q3'])
            else:
                HAM_quali.append(100)

df_2008 = pd.DataFrame(list(zip(lead_score, HAM_score, MAS_score, lead_quali08, HAM_quali, MAS_quali)), columns = ['Leading Score', 'HAM Score', 'MAS Score', 'Fastest Quali', 'HAM Quali', 'MAS Quali'])
df_2008.head(18)



```
